```python
train = pd.get_dummies(train, columns = ['Type'])
train
```

```python
train = train.rename(columns={'Machine failure': 'Machine_failure'})
```

```python
import optuna 
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = train.drop(columns = 'Machine_failure')
y = train.Machine_failure
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
def obj(trial):
    n_estimators = trial.suggest_int('n_estimators', 200, 900, log = True)
    max_depth = trial.suggest_int('max_depth', 2, 12, log = True)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 25)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 25)
    max_samples = trial.suggest_float('max_samples', 0.1, 0.99)
    
    model = RandomForestClassifier(
        n_estimators = n_estimators,
        max_depth = max_depth,
        min_samples_split = min_samples_split,
        min_samples_leaf = min_samples_leaf,
        max_samples = max_samples,
        bootstrap = True,
        oob_score = True,
        random_state = 42,
        n_jobs = -1
    )
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)
```

```python
study = optuna.create_study(direction = 'maximize')
study.optimize(obj, n_trials = 100)
```

```python
model = RandomForestClassifier(**study.best_params)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_pred, y_test))
```
